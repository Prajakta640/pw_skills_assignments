```python
#Question4- How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [2, 3, 4, 5, 6],
        'C': [5, 4, 3, 2, 1],
        'D': [1, 3, 5, 7, 9]}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Title
plt.title("Correlation Matrix Heatmap")

# Show plot
plt.show()
```

#Question2- How does broadcasting work in NumPy
**Answer**- Broadcasting in NumPy allows array operations without explicit looping by automatically expanding smaller arrays to match the shape of larger ones. This makes computations more efficient and memory-friendly.

Basic Broadcasting Rules
Matching Dimensions: If two arrays have the same shape, element-wise operations occur normally.
Adding a Dimension (Singleton Expansion): If one array has a dimension of 1, it can be stretched to match the other.
Dimension Mismatch: If dimensions dont match and cannot be expanded, NumPy throws an error.

Key Benefits of Broadcasting
 Removes Explicit Loops  Makes operations more Pythonic and faster.
 Memory Efficient  Avoids unnecessary array duplication.
 Optimized Performance  Uses low-level C implementations.

#Question12- What is the purpose of the describe() function in Pandas**

**Answer**- The describe() function in Pandas is used to generate summary statistics of a DataFrame or a Series. It provides a quick overview of the distribution and key statistical metrics for numeric data, such as the mean, standard deviation, min, max, and percentiles.

Key Statistics Produced by describe():
1. Count: The number of non-null values in each column.
2. Mean: The average of the values in each column.
3. Standard deviation (std): Measures the spread or dispersion of the data.
4. Min: The minimum value in each column.
5. 25th percentile (25%): The value below which 25% of the data falls.
6. 50th percentile (50%): The median value of the data.
7. 75th percentile (75%): The value below which 75% of the data falls.
8. Max: The maximum value in each column.

#Question17- Explain the difference between apply() and map() in Pandas**

#Question20- What is the role of a pivot table in Pandas**

**Answer**- A pivot table in Pandas is a powerful tool for summarizing and aggregating data. It allows you to reshape and summarize data in a way that makes it easier to analyze and understand patterns, relationships, and insights. Essentially, a pivot table groups data by one or more columns and performs an aggregation (e.g., sum, mean, count) on the remaining columns.

The pivot_table() function in Pandas provides a flexible way to create these summaries by specifying:

Values: The column(s) to aggregate.
Index: The column(s) to group by, forming the rows of the pivot table.
Columns: The column(s) to use as the columns of the pivot table (optional).
Aggregation Function: The function (e.g., sum, mean, count) used to aggregate the data.

Efficient Summary: They allow you to quickly summarize and aggregate large datasets in a compact format.

Better Data Insights: Pivot tables help to uncover trends, patterns, and relationships between different categories in your data.

Flexibility: You can easily manipulate, filter, and customize pivot tables for specific analyses.
